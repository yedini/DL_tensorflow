- sentence를 word 단위로 Tokenization: 한 문장의 단어들을 sequence로 보고 sentence를 word 단위로 분해하는 것.
- sequence들을 RNN 모델에 입력해서 positive 또는 negative로 classification

image.png

- Embedding layer: 토큰을 RNN 모델에 넣기 위해 numeric data로 바꿔주는 층. 

#### Example: word sentiment classification

```python
# setup
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import Sequential, Model
from tensorflow.keras.preprocessing.sequence import pad_sequences
%matplotlib inline

print(tf.__version__)
```

```python
# example data
words = ['good', 'bad', 'worse', 'so good'] # 예제 word
y_data = [1,0,0,1]  # 예제 word에 대한 긍정/부정 label

# creating a token dictionary
# word를 sequence로 바꿨을 때 각 길이가 달라서 길이를 맞추기 위해 <pad>를 추가해줌.
char_set = ['<pad>'] + sorted(list(set(''.join(words))))
idx2char = {idx : char for idx, char in enumerate(char_set)}
char2idx = {char : idx for idx, char in enumerate(char_set)}

print(char_set)
print(idx2char)
print(char2idx)
```

```python
# converting sequence of tokens to sequence of indices
x_data = list(map(lambda word : [char2idx.get(char) for char in word], words))
x_data_len = list(map(lambda word : len(word), x_data))

print(x_data) # 각 word가 integer index의 sequence로 변환됨.
print(x_data_len)
```

```python
# padding the sequence of indices
#  각 sequence가 max_sequence만큼의 길이를 갖도록 빈 값을 padding으로 채움. => 0으로 채워짐. 
max_sequence = 10
x_data = pad_sequences(sequences = x_data, maxlen = max_sequence,
                       padding = 'post', truncating = 'post')

# checking data
print(x_data)
print(x_data_len)
print(y_data)
```

```python
# creating simple rnn for "many to one" classification
input_dim = len(char2idx)
output_dim = len(char2idx)
one_hot = np.eye(len(char2idx))
hidden_size = 10
num_classes = 2

model = Sequential()
# layers.Embedding: One-hot vector를 사용함.
# mask_zero=True: 전처리 단계에서 0으로 padding된 값을 연산에서 제외.
# trainable=False: One-hot vector를 train하지 않음.
model.add(layers.Embedding(input_dim=input_dim, output_dim=output_dim,
                           trainable=False, mask_zero=True, input_length=max_sequence,
                           embeddings_initializer=keras.initializers.Constant(one_hot)))
model.add(layers.SimpleRNN(units=hidden_size)) #sequence의 마지막 토큰을 input으로 받아 처리함.
model.add(layers.Dense(units=num_classes))
```

```python
model.summary()
```

- embedding의 output shape: (data dimension, max sequence, input dimension)
- simple rnn: RNN에서 설정한 hidden_size만큼의 벡터로 처리
- dense: classification

```python
# creating loss function
def loss_fn(model, x, y):
    return tf.reduce_mean(tf.keras.losses.sparse_categorical_crossentropy(
        y_true=y, y_pred=model(x), from_logits=True))

# creating an optimizer
lr = .01
epochs = 30
batch_size = 2
opt = tf.keras.optimizers.Adam(learning_rate = lr)
```

```python
# generating data pipeline
tr_dataset = tf.data.Dataset.from_tensor_slices((x_data, y_data))
tr_dataset = tr_dataset.shuffle(buffer_size = 4)
tr_dataset = tr_dataset.batch(batch_size = batch_size)

print(tr_dataset)
```

```python
# training
tr_loss_hist = []

for epoch in range(epochs):
    avg_tr_loss = 0
    tr_step = 0
    
    for x_mb, y_mb in tr_dataset:
        with tf.GradientTape() as tape:
            tr_loss = loss_fn(model, x=x_mb, y=y_mb)
        grads = tape.gradient(target=tr_loss, sources=model.variables)
        opt.apply_gradients(grads_and_vars=zip(grads, model.variables))
        avg_tr_loss += tr_loss
        tr_step += 1
    else:
        avg_tr_loss /= tr_step
        tr_loss_hist.append(avg_tr_loss)
    
    if (epoch + 1) % 5 ==0:        
        print('epoch : {:3}, tr_loss : {:.3f}'.format(epoch + 1, avg_tr_loss.numpy()))
```

```python
# Checking performance
yhat = model.predict(x_data)
yhat = np.argmax(yhat, axis=-1)
print('acc : {:.2%}'.format(np.mean(yhat == y_data)))
plt.plot(tr_loss_hist)
```
